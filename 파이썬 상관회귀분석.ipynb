```python
#회귀분석 실시
from scipy import stats
lm = stats.linregress(df.V7, df.V9) 

#기울기, 절편, 상관계수, 오류지수, 표준오차
lm #p-value가 2.405까지 나와서 무의미하다고 나온다..!

#회귀식 : y = 절편 + 기울기*x 
#y = 272 + 6.28 * x
```
